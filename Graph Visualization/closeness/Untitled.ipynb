```python
import networkx as martina
import matplotlib.pyplot as plt
 
g=martina.Graph()



with open("go.txt") as f:
    for line in f:
        int_list = [int(x) for x in line.split()]  
        From=int_list[0]
        To=int_list[1]
        W=int_list[2]
        g.add_edge(From,To)
        g.add_weighted_edges_from([(From,To,W)])
        
f.close()

with open("marina.txt") as K:
    for line in K:
        int_list = [float(x) for x in line.split()]
nod_list = [int(x*1200) for x in int_list]

martina.draw(g,with_labels=True,node_size=nod_list)
plt.draw()
plt.show()
```
