```python
import networkx as martina
import matplotlib.pyplot as plt

g=martina.Graph()

with open("go.txt") as f:
    for line in f:
        int_list = [int(x) for x in line.split()]
        From=int_list[0]
        To=int_list[1]
        W=int_list[2]
        #g.add_edge(From,To)
        g.add_weighted_edges_from([(From,To,W)])
        
f.close()


nod =g.degree()
nod_list = [int(x[1]*600) for x in nod]

martina.draw(g,with_labels=True,node_size=nod_list)
plt.draw()
plt.show()
```
